```python
from collections import Counter
```

```python
arrival_date_month = data1["arrival_date_month"] 

arrivaldatemonth = Counter()
```

```python
for arrivaldate in arrival_date_month:
    arrivaldatemonth.update(arrivaldate.split(';'))
    
print(arrivaldatemonth.most_common())
```

```python
month=[]
qty = []

for item in arrivaldatemonth.most_common():
    month.append(item[0])
    qty.append(item[1])

print(month)
print(qty)
```

```python

plt.bar(month, qty , color="b")

plt.style.use("seaborn")
plt.title("Month of The Most Visitor Come")
plt.xlabel("Number of Visitors")
plt.ylabel("Month")
plt.tight_layout()

plt.savefig("D:/Most.png")
plt.show()
```

```python
time = data1["Time"] 

SpecTime = Counter()
```

```python
for specttime in time:
    SpecTime.update(specttime.split(';'))
    
print(SpecTime.most_common())
```

```python
Time=[]
Spec = []

for item in SpecTime.most_common():
    Time.append(item[0])
    Spec.append(item[1])

print(Time)
print(Spec)
```

```python
colors = ["green","red","yellow"]
explode = [0,0,0]

plt.style.use("seaborn")
labels= ["Short", "Long", "Medium"]
plt.pie(Spec , labels=labels , autopct="%1.2f%%", colors=colors, explode=explode)
plt.title(" Lead Time Categories")
plt.tight_layout()

plt.savefig("D:/LeadTimeCategories.png")
plt.show()
```
